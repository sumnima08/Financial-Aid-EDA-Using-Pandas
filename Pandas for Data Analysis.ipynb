```python
#Pandas for Data Analysis
#Import pandas
import pandas as pd
```

```python
#Find working directory
import os
os.getcwd()
```

```python
#Read a CSV file
df1 = pd.read_csv('grades.csv')
```

```python
#View the first 10 rows of df1
print(df1.head(10))
```

```python
#View the bottom 10 rows of df1
print(df1.tail(10))
```

```python
#Summary Statistics
df1.describe()
```

```python
#Check for duplicate rows
duplicate_count = df1.duplicated().sum()
print(duplicate_count)

#duplicates = df1. duplicated()
#print(duplicates)

#View instances of the duplicated rows
duplicated_rows = df1[df1.duplicated()]
duplicated_rows
```

```python
#Drop duplicate rows
df1.drop_duplicates(inplace=True)
```

```python
#Count missing values and drop rows with missing values
print(df1.isnull().sum())

df1.dropna(inplace=True) # "inplace=True" modifies the existing data frame
print(df1.isnull().sum())
```

```python
#Creating a new column "honour" 
def h_gpa(x):
    if x>3.5:
        return 'Yes'
    else:
        return 'No'
df1['Honour Student']= df1['GPA'].map(h_gpa)
df1
```

```python
#Read Excel File
df2 = pd.read_excel('tuition.xlsx')
```

```python
#Count missing values and drop rows with missing values
print(df2.isnull().sum())

df2.dropna(inplace=True) #"inplace=True" modifies the existing data frame
print(df2.isnull().sum())
```

```python
#Check for duplicate rows
duplicate_count = df2.duplicated().sum()
print(duplicate_count)

#duplicates = df2.duplicated()
#print(duplicates)

#View instances of the duplicated rows
duplicated_rows = df2[df2.duplicated()]
duplicated_rows
```

```python
#Drop duplicate rows
df2.drop_duplicates(inplace=True)
```

```python
#Map with a function
df2['Cost'] = df2['Tuition']- df2['Financial Aid']

#Display the Cost
print(df2)
```

```python
#Mering data frames using Inner Join on the 'ID' column

df3 = pd.merge(df1, df2, on='ID', how='inner')
print(df3)
```

```python
#Count missing values and drop rows with missing values
print(df3.isnull().sum())

df3.dropna(inplace=True) # 'inplace=True' modifies the existing the data frame
print(df3.isnull().sum())
```

```python
#Group the students by major and display the mean GPA 
df_major = df3.groupby('Major')
df_major['GPA']
print(df_major['GPA'].mean())
```

```python
#Display net cost by major
df_major['Cost']

df_major['Cost'].mean()
```

```python
#Filtering everyone with less than 2.3 GPA from df3
df_filter = df3[df3['GPA'] > 2.3]
print(df_filter)
```

```python
#Sorting ID by descending order
df_desc = df3.sort_values('ID', ascending= False)
print(df_desc)
```

```python
#Exporting df3 as a CSV 
df3.to_csv('output.csv',index= False)
```

```python
#Exporting df3 as an excel file
df3.to_excel('output.xlsx', index= False)
```
